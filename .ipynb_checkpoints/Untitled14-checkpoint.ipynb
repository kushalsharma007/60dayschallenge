```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load the dataset
dataset=pd.read_csv("C://Users//Acer//Downloads//dataset.csv")
dataset.head()
```

```python
#define x and y
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,1].values
```

```python
#lets see the value for x and y
X
```

```python
Y
```

```python
#split the dataset into training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.3, random_state=0)
```

```python
#see the values together
X_train,X_test,Y_train,Y_test
```

```python
#Regression
from sklearn.linear_model import LinearRegression
#make a function for regression
reg= LinearRegression()
#Lets fit the model with split dataset
reg.fit(X_train,Y_train)
```

```python
#prediction
Y_predict=reg.predict(X_test)
Y_predict
```

```python
#now lets display graph
plt.scatter(X_train, Y_train, color='red')
plt.plot(X_train,reg.predict(X_train), color='blue')
plt.title("Linear Regression salary vs experience",size=18)
plt.xlabel("Year of employees",size=15)
plt.ylabel("Salary of employees",size=15)
plt.show()
```

```python
plt.scatter(X_test, Y_test, color='red')
plt.plot(X_test,reg.predict(X_test), color='blue')
plt.title("Linear Regression salary vs experience",size=18)
plt.xlabel("Year of employees",size=15)
plt.ylabel("Salary of employees",size=15)
plt.show()
```
