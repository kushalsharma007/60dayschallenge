```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resize
plt.figure(figsize=(5,3), dpi=180)
#plot line 1
#shorthand notation
plt.plot(x,y, 'b^--', label='2x')

#line number two
x2 = np.arange(0,4,0.5)

plt.plot(x2[:5], x2[:5]**2,'r'  ,label='X^2')
plt.plot(x2[4:], x2[4:]**2,'r--')
plt.title('First Graph', fontdict={'fontsize':22})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])

plt.legend()

plt.savefig('mygraph.png', dpi=300)
plt.show()
```

```python
labels = ['A','B','C']
values=[4,6,8]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('*')
bars[2].set_hatch('o')

plt.figure(figsize=(4,8))

plt.show()
```
