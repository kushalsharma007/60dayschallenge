```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Binary cross-entropy loss
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15  # to avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
# Perceptron model
class Perceptron:
    def __init__(self, input_dim):
        self.weights = np.random.randn(input_dim)
        self.bias = 0
    
    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return sigmoid(linear_output)
    
    def train(self, X, y, learning_rate=0.1, epochs=100):
        losses = []
        for epoch in range(epochs):
            y_pred = self.predict(X)
            loss = binary_cross_entropy(y, y_pred)
            losses.append(loss)
            # Gradient descent
            error = y_pred - y
            dW = np.dot(X.T, error * sigmoid_derivative(y_pred)) / X.shape[0]
            dB = np.mean(error * sigmoid_derivative(y_pred))
            
            self.weights -= learning_rate * dW
            self.bias -= learning_rate * dB
            
            print(f'Epoch {epoch + 1}/{epochs} - Loss: {loss}')
        
        return losses
```

```python
# Generate dummy data
np.random.seed(42)
X = np.random.randn(200, 2)
y = (np.sum(X, axis=1) > 0).astype(int)  # simple linear boundary

# Initialize and train perceptron
perceptron = Perceptron(input_dim=2)
losses = perceptron.train(X, y, learning_rate=0.1, epochs=100)
```

```python
# Plot loss over epochs
plt.figure(figsize=(12, 5))

# Plot data points
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', alpha=0.7)
plt.title('Data Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
# Plot loss reduction over epochs
plt.subplot(1, 2, 2)
plt.plot(range(1, 101), losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Reduction Over Epochs')

plt.tight_layout()
plt.show()
```

```python
# Plot decision boundary
xx, yy = np.mgrid[-3:3:.01, -3:3:.01]
grid = np.c_[xx.ravel(), yy.ravel()]
probs = perceptron.predict(grid).reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, probs, levels=[0, 0.5, 1], cmap="bwr", alpha=0.6)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr", edgecolor="k", alpha=0.7)
plt.title('Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
