```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Define the neural network class
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, learning_rate):
        # Initialize weights with a large value
        self.weights_input_hidden = np.random.randn(input_size, hidden_size) * 10
        self.weights_hidden_output = np.random.randn(hidden_size, output_size) * 10
        self.learning_rate = learning_rate

    def forward(self, X):
        self.hidden_input = np.dot(X, self.weights_input_hidden)
        self.hidden_output = sigmoid(self.hidden_input)
        self.final_input = np.dot(self.hidden_output, self.weights_hidden_output)
        self.final_output = sigmoid(self.final_input)
        return self.final_output

    def backward(self, X, y, output):
        self.output_error = y - output
        self.output_delta = self.output_error * sigmoid_derivative(output)
        
        self.hidden_error = self.output_delta.dot(self.weights_hidden_output.T)
        self.hidden_delta = self.hidden_error * sigmoid_derivative(self.hidden_output)
        
        # Update weights with a high learning rate
        self.weights_hidden_output += self.hidden_output.T.dot(self.output_delta) * self.learning_rate
        self.weights_input_hidden += X.T.dot(self.hidden_delta) * self.learning_rate

    def train(self, X, y, epochs):
        losses = []
        for epoch in range(epochs):
            output = self.forward(X)
            self.backward(X, y, output)
            loss = np.mean(np.square(y - output))
            losses.append(loss)
            if epoch % 100 == 0:
                print(f'Epoch {epoch}, Loss: {loss}')
                self.plot_predictions(X, y, epoch)
        return losses

    def plot_predictions(self, X, y, epoch):
        output = self.forward(X)
        plt.scatter(X[:, 0], X[:, 1], c=y.flatten(), cmap='viridis', label='Actual')
        plt.scatter(X[:, 0], X[:, 1], c=output.flatten(), cmap='coolwarm', marker='x', label='Predicted')
        plt.title(f'Epoch {epoch}')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.legend()
        plt.show()
```

```python
# Generate some random data for training
np.random.seed(42)
X = np.random.rand(100, 2)
y = np.array([[1] if x[0] + x[1] > 1 else [0] for x in X])
```

```python
# Initialize and train the neural network with a high learning rate
input_size = 2
hidden_size = 10
output_size = 1
learning_rate = 1.0  # High learning rate

nn = NeuralNetwork(input_size, hidden_size, output_size, learning_rate)
losses = nn.train(X, y, epochs=1000)
```

```python
# Plot the loss over epochs
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Over Epochs (Exploding Gradients)')
plt.show()
```
