```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to initialize parameters
def initialize_parameters(input_size, hidden_size, output_size):
    np.random.seed(1)
    W1 = np.random.randn(hidden_size, input_size) * 0.01
    b1 = np.zeros((hidden_size, 1))
    W2 = np.random.randn(output_size, hidden_size) * 0.01
    b2 = np.zeros((output_size, 1))
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

```python
# Function to perform the forward pass
def forward_pass(X, parameters):
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    
    Z1 = np.dot(W1, X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = Z2  # Linear activation for regression
    
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, cache
```

```python
# Function to compute the cost
def compute_cost(A2, Y):
    m = Y.shape[1]
    cost = (1 / (2 * m)) * np.sum((A2 - Y) ** 2)
    return cost
```

```python
# Function to perform the backward pass
def backward_pass(parameters, cache, X, Y):
    m = X.shape[1]
    
    W2 = parameters['W2']
    A1 = cache['A1']
    A2 = cache['A2']
    
    dZ2 = A2 - Y
    dW2 = (1 / m) * np.dot(dZ2, A1.T)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)
    
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = dA1 * (1 - np.power(A1, 2))
    dW1 = (1 / m) * np.dot(dZ1, X.T)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}
    
    return grads
```

```python
# Function to update the parameters
def update_parameters(parameters, grads, learning_rate):
    parameters['W1'] = parameters['W1'] - learning_rate * grads['dW1']
    parameters['b1'] = parameters['b1'] - learning_rate * grads['db1']
    parameters['W2'] = parameters['W2'] - learning_rate * grads['dW2']
    parameters['b2'] = parameters['b2'] - learning_rate * grads['db2']
    
    return parameters
```

```python
# Function to train the neural network
def train(X, Y, input_size, hidden_size, output_size, num_iterations, learning_rate):
    parameters = initialize_parameters(input_size, hidden_size, output_size)
    
    for i in range(num_iterations):
        A2, cache = forward_pass(X, parameters)
        cost = compute_cost(A2, Y)
        grads = backward_pass(parameters, cache, X, Y)
        parameters = update_parameters(parameters, grads, learning_rate)
        
        if i % 100 == 0:
            print(f"Iteration {i}, Cost: {cost}")
    
    return parameters
```

```python
#functions to make predictions
def predict(X, parameters):
    A2, _ = forward_pass(X, parameters)
    return A2

# Sample data (X, Y)
np.random.seed(1)
X = np.random.randn(2, 500)
Y = np.sin(X[0,:]) + np.cos(X[1,:])
Y = Y.reshape(1, 500)

# Training the neural network
input_size = X.shape[0]
hidden_size = 10
output_size = 1
num_iterations = 1000
learning_rate = 0.01

parameters = train(X, Y, input_size, hidden_size, output_size, num_iterations, learning_rate)

# Predicting
predictions = predict(X, parameters)
```

```python
# Plotting the results
plt.scatter(X[0, :], Y, color='red', label='Original data')
plt.scatter(X[0, :], predictions, color='blue', label='Predicted data')
plt.title('Original vs Predicted data')
plt.xlabel('X[0]')
plt.ylabel('Y')
plt.legend()
plt.show()
```
