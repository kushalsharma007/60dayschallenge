```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#visualize some data points
x = [2, 3, 8, 2, 1, 9, 12 , 4, 8, 10]
y = [18, 16, 21, 14, 13, 22, 21, 19, 18, 18]

plt.scatter(x, y)
plt.show()
```

```python
#utilize elbow method to visualize the inertia for different values of k
data = list(zip(x, y))
inertias = []
for i in range(1,9):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
plt.plot(range(1,9), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
#it shows 2 is a good value of k
#retrain and visualize the result
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
