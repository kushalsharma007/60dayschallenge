```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate synthetic dataset
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.array([1 if x[0] + x[1] > 0 else 0 for x in X])
```

```python
def perceptron(X, y, epochs=1000, lr=0.01):
    #insert bias term
    X = np.insert(X, 0, 1, axis=1)
    weights = np.random.rand(X.shape[1])
    
    for _ in range(epochs):
        for i in range(len(y)):
            y_hat = step(np.dot(X[i], weights))
            weights = weights + lr * (y[i] - y_hat) * X[i]

    return weights[0], weights[1:]
```

```python
def step(z):
    return 1 if z > 0 else 0
```

```python
intercept_, coef_ = perceptron(X, y)
print("Coefficients:",coef_)
print("Intercept:", intercept_)
```

```python
#plot decision boundary
m = -(coef_[0] / coef_[1])
b = (intercept_ / coef_[1])
x_input = np.linspace(min(X[:,1]), max(X[:,1]), 100)
y_input = m* x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color="red", linewidth=2)
plt.scatter(X[:,0], X[:,1], c=y, cmap="autumn", s=100)
plt.title("Perceptron Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.ylim(min(X[:,1]) -1, max(X[:,1]) +1)
plt.show()
```
